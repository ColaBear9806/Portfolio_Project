```python
import pytesseract
import cv2
import os
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
#이미지를 불러와 gray 스케일로 변환해주기.

image = cv2.imread('sample.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#pytesseract에서는 numpy array를 읽지 못하고 file을 읽기 때문에 os로 파일을 불러들여야 함.
filename = "{}.png".format(os.getpid())
cv2.imwrite(filename, gray)

tessdata_dir_config = r'--tessdata-dir "/content/tessdata"'

#pytesseract의 image to string 사용.


text = pytesseract.image_to_string(Image.open(filename), lang = 'eng')
#text = pytesseract.image_to_string(Image.open(filename), config=tessdata_dir_config, lang = 'kor')
os.remove(filename)


print(text)
cv2_imshow(image)
```
